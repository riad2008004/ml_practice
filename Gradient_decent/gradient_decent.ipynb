```python
#Gradient Decent
```

```python
from sklearn.linear_model import SGDRegressor
```

```python
gd=SGDRegressor(learning_rate='constant', eta0=0.01, max_iter=1000)
```

```python
gd.fit(df[['x']],df[['y']])
```

```python
gd.score(df[['x']],df[['y']])
```

```python
from sklearn.preprocessing import StandardScaler

```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(df[['x']])
```

```python
gd.fit(x_scaled,df[['y']])
```

```python
gd.score(x_scaled,df[['y']])
```
