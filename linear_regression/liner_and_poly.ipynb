```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
linear_data_frame= pd.read_csv('linear_data.csv')
```

```python
linear_data_frame.head()
```

```python
x_linear=linear_data_frame[['x']]
y_linear=linear_data_frame[['y']]
```

```python
non_linear_dataframe=pd.read_csv('nonlinear_data.csv')
```

```python
x_nonlinear=non_linear_dataframe[['x']]
y_nonlinear=non_linear_dataframe[['y']]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.scatter(x_linear,y_linear, color='blue',label='Data linear')
plt.xlabel=('X')
plt.ylabel=('Y')
plt.title=('Linear Relationship')
plt.legend()

plt.subplot(1, 2, 2)
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.xlabel=('X')
plt.ylabel=('Y')
plt.title=('Non-linear Relationship')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
linear_data_frame.corr()
```

```python
non_linear_dataframe.corr()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg_1=LinearRegression()
```

```python
reg_1.fit(x_linear,y_linear)
```

```python
reg_1.score(x_linear,y_linear)
```

```python
plt.scatter(x_linear, y_linear, color='blue', label='Data with Linear Relationship')
plt.plot(x_linear, reg_1.predict(x_linear), color='red', label='Best fit line')
plt.xlabel=('X')
plt.ylabel=('Y')
plt.title=('Linear Relationship')
plt.legend()
```

```python
reg_2=LinearRegression()
```

```python
reg_2.fit(x_nonlinear,y_nonlinear)
```

```python
reg_2.score(x_nonlinear,y_nonlinear)
```

```python
plt.scatter(x_nonlinear, y_nonlinear, color='green', label='Data with Non-linear Relationship')
plt.plot(x_nonlinear, reg_2.predict(x_nonlinear), color='red', label='Best fit line')
plt.xlabel=('X')
plt.ylabel=('Y')
plt.title=('Non-linear Relationship')
plt.legend()
```
