```python
import pandas as pd
```

```python
df=pd.DataFrame()
```

```python
df=pd.read_csv('C:/Users/Asus/Downloads/merged_file.csv')
```

```python
df
```

```python
df_time_pitch = df[['time', 'Pitch']]
```

```python
df_time_pitch
```

```python
import matplotlib.pyplot as plt

time = df['time']
pitch = df['Pitch']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, pitch, label='Pitch', color='blue', linewidth=1.5)

# Customize the plot
plt.title('Pitch Over Time', fontsize=16)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Pitch (degrees)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df_time_pitch['Pitch']=df_time_pitch['Pitch']-105
```

```python
df_time_pitch
```

```python
time = df_time_pitch['time']
pitch = df_time_pitch['Pitch']

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(time, pitch, label='Pitch', color='blue', linewidth=1.5)

# Customize the plot
plt.title('Pitch Over Time', fontsize=16)
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Pitch (degrees)', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df_time_pitch.to_csv('normal_pitch_over_time.csv', index=False)
```
